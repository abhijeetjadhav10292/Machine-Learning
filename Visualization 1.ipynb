
### Figsize, labels and title

```python
area = [600,800,1000,1200,1500]
price = [15,18,25,35,40]
```

```python
plt.figure(figsize=(7,7))
plt.plot(area,price)
plt.title("Area vs Price")
plt.xlabel("Area in sq ft")
plt.ylabel("Price in lakh")
plt.show()
```

### xlim and ylim to define x and y range

```python
plt.figure(figsize=(7,7))
plt.plot(area,price)
plt.title("Area vs Price")
plt.xlabel("Area in sq ft")
plt.ylabel("Price in lakh")
plt.xlim((400,2000))
plt.ylim((10,50))
plt.show()
```

### xticks and yticks to rotate labels

```python
plt.figure(figsize=(7,7))
plt.plot(area,price)
plt.title("Area vs Price")
plt.xlabel("Area in sq ft")
plt.ylabel("Price in lakh")
plt.xticks(rotation=45)
plt.yticks(rotation=90)
plt.show()
```

### Multiple plots

```python
x = np.arange(1,20)
y1 = np.sqrt(x)
y2 = np.log(x)
```

```python
plt.figure(figsize=(7,7))
plt.plot(x,y1,color="blue")
plt.plot(x,y2,color="red")
plt.legend(labels=["square root","log"])
plt.show()
```

## 2) Histograms 

```python
df = pd.read_csv("Automobile_data.csv")
```

```python
df.info()
```

```python
df.describe()
```

```python
plt.figure(figsize=(7,7))
plt.hist(df["width"], bins=10)
plt.title("Width frequency distribution")
plt.ylabel("frequency")
plt.xlabel("width")
plt.show()
```

```python
# Using pandas
plt.figure(figsize=(7,7))
df["width"].hist(bins=10)
plt.title("Width frequency distribution")
plt.ylabel("frequency")
plt.xlabel("width")
plt.show()
```

```python
# Plot mean and median

width_mean = df["width"].mean()
width_median = df["width"].median()

plt.figure(figsize=(7,7))
df["width"].hist(bins=10)
plt.title("Width frequency distribution")
plt.axvline(width_mean,color="red")
plt.axvline(width_median,color="yellow")
plt.ylabel("frequency")
plt.xlabel("width")
plt.show()
```

## 3) Bar plot

```python
plt.figure(figsize=(9,9))
df["make"].value_counts().plot(kind="bar")
plt.title("Make count bar plot")
plt.ylabel("Count")
plt.xlabel("Make")
plt.show()
```

```python
# horizontal plot
plt.figure(figsize=(9,7))
df["make"].value_counts().plot(kind="barh")
plt.title("Make count bar plot")
plt.ylabel("Make")
plt.xlabel("Count")
plt.show()
```

## 4) Pie chart

```python
plt.figure(figsize=(7,7))
df["body-style"].value_counts().plot.pie()
plt.title("Body-style count pie chart")
plt.show()
```

```python
# Autopct and explode
plt.figure(figsize=(7,7))
df["body-style"].value_counts().plot.pie(autopct="%1.1f%%", explode=(0,0,0,0.1,0))
plt.title("Body-style count pie chart")
plt.show()
```

## 5) Scatter plot 

```python
# Correlation
df.corr()
```

```python
plt.figure(figsize=(7,7))
plt.scatter(x=df["width"],y=df["price"])
plt.title("Width vs Price Scatter plot")
plt.ylabel("price")
plt.xlabel("width")
plt.show()
```

```python
plt.figure(figsize=(7,7))
plt.scatter(x=df["height"],y=df["price"],color="red")
plt.title("Height vs Price Scatter plot")
plt.ylabel("price")
plt.xlabel("height")
plt.show()
```

```python
plt.figure(figsize=(7,7))
plt.scatter(x=df["city-mpg"],y=df["price"])
plt.title("City-mpg vs Price Scatter plot")
plt.ylabel("price")
plt.xlabel("city-mpg")
plt.show()
```
